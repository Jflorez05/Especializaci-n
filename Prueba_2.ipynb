## **Ejercicio 2** (30%)

Importe el dataset `athletes` del repositorio del curso y realice las siguientes tareas:

1. Importarlo como nombre athletes

```python
import pandas as pd
import numpy as np

athletes = pd.read_csv('https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv')
athletes.head()
```

2. Filtre el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla.

```python
# Filtrado del dataset con unicamente los atletas que han ganado al menos una medalla.
athletes_medal = athletes.loc[(athletes.gold > 0) | (athletes.silver > 0) | (athletes.bronze > 0) ]
athletes_medal
```

3. Calcule la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

```python
# Cálculo del indice de masa por fila con una función

def mass_index(row, weight, height):

    """
    Calculates the Body Mass Index (BMI) for a specific row of a DataFrame.

    Parameters:
        - row: A specific row of the DataFrame.
        - weight: The name of the column containing athletes' weights.
        - height: The name of the column containing athletes' heights.

    Returns:
        The calculated Body Mass Index (BMI) for the given row.
    """
    return row[weight]/((row[height])**2)

# Aplica la función mass_index a cada fila de athletes_medal y almacena el resultado en una nueva columna llamada 'mass_index'
athletes_medal['mass_index'] =athletes_medal.apply(mass_index,
                                        axis = 1,
                                        args = [ 'weight', 'height'])

athletes_medal
```

4. Agrupe los datos por calculando la media por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

**Respuesta 2.4:** El deporte que tiene el mayor peso promedio es el **Basketball.**

```python
# Agrupación de los datos calculando la media de las carécteristicas por deporte y seleccionar deporte con mayor peso promedio
athletes_medal.groupby('sport').mean().sort_values('weight',ascending= False)
```

5. ¿Cuál es el pais con más participación femenina?

**Respuesta 2.5**: El pais con una mayor participación femenina es **Estados unidos**.

```python
# Cuál es el pais con la mayor participación femenina

athletes_medal.groupby(by =['nationality', 'sex']).sex.count().sort_values(ascending=False)
```
